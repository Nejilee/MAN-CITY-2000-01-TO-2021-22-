```python

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('/home/neji/Downloads/ManCity.xlsx')
data.plot(x="SEASON", y="WINS", kind="bar", title="MANCHESTER CITY WINS FROM 2000/01-2021/22(EPL)")
plt.ylabel("NUMBER OF WINS")
plt.xlabel("PER SEASON")
plt.show()
```

```python
import matplotlib.pyplot as plt
data = pd.read_excel('/home/neji/Downloads/ManCity.xlsx')
data.plot(x="SEASON", y="END POSITION", kind="bar", color="lightseagreen", rot=90, title="MANCHESTER CITY WINS FROM 2000/01-2021/22(EPL)")
plt.yticks([1, 5, 10, 15, 20])
plt.ylabel("SEASON END POSITION")
plt.show()
```
